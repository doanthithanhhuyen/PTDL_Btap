```python
#xác định tên các cột
data.columns
```

```python
#đổi tên chỉ 1 cột xđ
data.rename(columns = {'Place': 'Địa điểm'}, inplace = True)
data
```

```python
#đổi tên toàn bộ các cột
data.columns = ['Địa điểm', 'Mã SP', 'Tên SP', 'Đơn vị', 'Tên ĐV', 'Tháng', 'Năm', 'Giá']
data
```

```python
#thêm cột mới vào cuối
data['Cột mới'] = np.nan
data
```

```python
#thêm 1 cột vào giữa các cột
data.insert(8,'chiết khấu (%)', np.random.randint(0,20, size=(data.shape[0], 1)))
display(data)
data.info()
```

```python
data.insert(5,'So Luong',10)
data
```

```python
#thêm cột giảm giá 10%
data['Giảm giá 10%'] = pd.Series('10%', index=data.index)
data
#hoặc
data.insert(11,'Giảm giá 2',pd.Series('12%', index=data.index))
data
```

```python
#Thêm 1 dòng mới
data=data.append({'Địa điểm':'NA','Mã SP':'RR','Tên SP':'Rice','Đơn vị':10,'Tên ĐV':'KG','Tháng':6,'Năm':2021,
                  'Giá':84.3785, 'Giảm giá 2':'12%','Giảm giá':'10%'},ignore_index=True)
data
```

```python
#Xoá một cột
del data['Cột mới']
data
```

```python
#xoá cột sử dụng phương thức pop
#trả về cột đã xoá
data.pop('Đơn vị')
```

```python
#xoá 1 cột sử dụng phương thức drop
# Trong đó axis = 1 là xóa cột; inplace = True xóa trục tiếp trên dữ liệu gốc mà không phải tạo bản sao
data.drop('Giảm giá', axis=1, inplace=True)
data
```

```python
#xoá các dòng
data.drop(1, axis = 0, inplace=True)
data
```

```python
# Xóa các dòng có chỉ số 7377 và 7379 sử dụng phương thức drop()
data.drop([7377,7379], inplace=True)
data
```
