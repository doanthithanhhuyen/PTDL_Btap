```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')
data.head()
```

```python
data.info()
```

```python
#Bộ dữ liệu gồm 541909 dòng và 8 cột
#Số hóa đơn,Mã kho, Mô tả,ngày mua,mã kh, quốc gia: định danh
#số lượng: định lượng
```

```python
#Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv

tg = data.loc[:,['Description','Quantity']]
tg.to_csv('OnlineRetail.csv')
tg
```

```python
#Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx
tg1 = data.iloc[0:1000]
tg1.to_excel('OnlineRetail.xlsx')
tg1
```

```python
#Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5
tg2 = data.loc[data.Quantity >= 10]
tg2 = pd.HDFStore('OnlineRetail.h5')
tg2['data'] = data
tg2['data']
```

```python
#Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json
tg3 = data.loc[1000:2000, ['Quantity', 'InvoiceDate', 'UnitPrice']]
tg3.to_json('OnlineRetail.json')
tg3
```

```python
#Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html
tg4 = data.loc[data.InvoiceNo == 536365]
tg4.to_html('OnlineRetail.html')
tg4
```
